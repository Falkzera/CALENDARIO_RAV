```python
import pandas as pd
from datetime import datetime, timedelta

xls = pd.ExcelFile("Agenda Atualizada 2025 - RAV.xlsx")

valid_sheets = [s for s in xls.sheet_names if s not in ["Fórmulas"]]

month_map = {
    "Janeiro": 1, "Fevereiro": 2, "Março": 3, "Abril": 4, "Maio": 5, "Junho": 6,
    "Julho": 7, "Agosto": 8, "Setembro": 9, "Outubro": 10, "Novembro": 11, "Dezembro": 12
}

events = []
event_id = 1

for sheet in valid_sheets:
    df = pd.read_excel(xls, sheet_name=sheet, header=None)
    month = month_map.get(sheet, 1)

    for col in range(df.shape[1]):
        for row in range(df.shape[0]):
            val = str(df.iloc[row, col]).strip() if pd.notna(df.iloc[row, col]) else ""
            if val.isdigit():  # dia encontrado
                dia = int(val)
                try:
                    header_row = row + 1
                    headers = [str(df.iloc[header_row, col+i]).strip().lower() if pd.notna(df.iloc[header_row, col+i]) else "" for i in range(5)]
                    if "hs" in headers[0] and "evento" in headers[4]:
                        for r in range(header_row+1, header_row+15):
                            hs = str(df.iloc[r, col]).strip() if pd.notna(df.iloc[r, col]) else ""
                            if not hs or hs.lower() == "nan": 
                                continue
                            if hs.isdigit(): 
                                break
                            condutor = str(df.iloc[r, col+1]).strip() if pd.notna(df.iloc[r, col+1]) else "A definir"
                            equipe = str(df.iloc[r, col+2]).strip() if pd.notna(df.iloc[r, col+2]) else ""
                            local = str(df.iloc[r, col+3]).strip() if pd.notna(df.iloc[r, col+3]) else ""
                            evento = str(df.iloc[r, col+4]).strip() if pd.notna(df.iloc[r, col+4]) else ""
                            
                            hora, minuto = 0, 0
                            try:
                                hs_norm = hs
                                if "h" in hs.lower() and ":" not in hs:
                                    hs_norm = hs.lower().replace("h", ":00")
                                if len(hs_norm) == 2: 
                                    hs_norm += ":00"
                                t = datetime.strptime(hs_norm.strip(), "%H:%M").time()
                                hora, minuto = t.hour, t.minute
                            except:
                                pass
                            
                            try:
                                dt = datetime(2025, month, dia, hora, minuto)
                                start = dt.isoformat()
                                end = (dt + timedelta(hours=1)).isoformat()
                                
                                events.append({
                                    "id": f"event_{event_id}",
                                    "title": evento[:40] if evento else "Evento",
                                    "start": start,
                                    "end": end,
                                    "condutor": condutor,
                                    "equipe": equipe,
                                    "local": local,
                                    "evento": evento,
                                    "backgroundColor": "#1f77b4",
                                    "borderColor": "#1f77b4",
                                    "textColor": "white"
                                })
                                event_id += 1
                            except:
                                continue
                except:
                    continue
```

```python
print(len(events))
for event in events:
    print(event)
# salvar e mjson
import json
with open('events.json', 'w', encoding='utf-8') as f:
    json.dump(events, f, ensure_ascii=False, indent=4)
```
